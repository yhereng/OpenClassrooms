### 2.2.2 Variable "circonférence_cm"

#### 2.2.2.1 identifiation et traitement des valeurs abérrantes

Une manière courante de définir les outliers est celle basée sur l'interquantile. Un point est alors considéré comme un outlier lorsque sa valeur est inférieure au premier quartile moins 1,5 fois l'interquantile ou lorsque sa valeur est supérieur au troisième quartile plus 1,5 fois l'interquantile. D'après le tableau précédent, on déduit que l'interquantile vaut 85cm, par conséquent, l'intervalle en dehors duquel la valeur peut être considérée commme outliers est [-97.5,242.5] qui dans notre cas revient à considérer comme outliers tout arbre ayant une circonférence supérieur à 242.5cm.

Cependant, en effectuant une recherche, on peut découvrir que le plus gros arbres de Paris mesure 8m de diamètre.Dans notre analyse, nous prendrons donc 800 pour valeur maximale

```python
df_outliers_circonference=df[df["circonference_cm"]>800]
```

```python
df_outliers_circonference.shape
```

il y a donc 77 outliers dans notre jeu de données.
Ces données abérrantes peuvent être dues à des erreurs de saisie, des erreurs de relevé ou des erreurs d'unité.
Ces outliers peuvent être ignorés, ou nous pouvons pousser l'analyse plus loin en essayant de comprendre ces valeurs abérrantes.Dans le cas présent, la quantité d'outlier étant très petite vis a vis de la quantité de données, nou décidons de simplement sortir les outliers de l'analyse.

```python
df=df[df["circonference_cm"]<=800]
```

```python
df.shape
```

**Encodage ordinal de la variable stade_developpement**

```python
df_imput.loc[df_imput.stade_developpement=='J','stade_dev_encode']=1
df_imput.loc[df_imput.stade_developpement=='JA','stade_dev_encode']=2
df_imput.loc[df_imput.stade_developpement=='A','stade_dev_encode']=3
df_imput.loc[df_imput.stade_developpement=='M','stade_dev_encode']=4
```

**Change 0 values to Nan**

```python
df_imput.stade_dev_encode.replace(0,np.nan,inplace=True)
```

**Imputation**

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=500)
knn_result=imputer.fit_transform(df_imput[{'circonference_cm','hauteur_m','stade_dev_encode'}])
df_imput.stade_dev_encode=knn_result[:,2]
```

**Stockage du résultat,reformattage et decodage**

```python
df_imput.stade_dev_encode=round(df_imput.stade_dev_encode)
df_imput.stade_dev_encode=df_imput.stade_dev_encode.astype(int)
#decodage
df_imput.loc[df_imput.stade_dev_encode==1,'stade_dev_imput']='J'
df_imput.loc[df_imput.stade_dev_encode==2,'stade_dev_imput']='JA'
df_imput.loc[df_imput.stade_dev_encode==3,'stade_dev_imput']='A'
df_imput.loc[df_imput.stade_dev_encode==4,'stade_dev_imput']='M'
```

**Critique des résultats**

```python
plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
g=sns.histplot(data=df_imput,
               x="circonference_cm",
               hue="stade_dev_imput",
               multiple="stack",
               binwidth=8,
               hue_order=['J','JA','A','M'])
plt.title('Disitribution de la variable circonference_cm tronquée à 300',fontsize=15,fontweight='bold')
plt.xlabel('Circonférence',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xlim(0,300)


plt.subplot(2,1,2)
g=sns.histplot(data=df_imput, x="circonference_cm",
               hue="stade_developpement",
               multiple="stack",
               binwidth=8,
               hue_order=['J','JA','A','M'])
plt.title('Disitribution de la variable circonference_cm tronquée à 300',fontsize=15,fontweight='bold')
plt.xlabel('Circonférence',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xlim(0,300)

plt.show()
```
