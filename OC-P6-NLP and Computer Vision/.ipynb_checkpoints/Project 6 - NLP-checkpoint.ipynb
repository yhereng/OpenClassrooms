```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(token_pattern=r'\w+')
```

```python
X = vectorizer.fit_transform(corpus)
```

```python
words =vectorizer.get_feature_names()
```

```python
sums = [i for i in X.sum(axis=0).tolist()[0]]
words_appearance = dict(zip(words,sums))
```

```python
sns.barplot(y = most_comun_words[0:20],
            x =[words_appearance[i] for i in most_comun_words[0:20]],
            orient='h')
plt.xticks(rotation=45)
plt.title('most commun words in the corpus')
plt.tight_layout()
```

## Stemming

```python
#create stemmer
stemmer = SnowballStemmer('english')

#create stem dataframe
df_stems = pd.DataFrame(index=df_text.index)

#iterate on every columns
for word in df_text.columns.values:
    stem = stemmer.stem(word)
    try:
        df_stems[stem]+=df_text[word] #if the counter exist add the new word
        
    except KeyError:
        df_stems[stem]=df_text[word] #if the counter does not exist,create it
    
```

```python
B = df_stems.groupby('stars_rat').sum()
T= B['text'].copy()
B.drop(columns=['text'],inplace=True)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10), facecolor='w', edgecolor='k')
axs = axs.ravel()
for rating in B.index:
    ax = axs[int(rating)-1]
    (B.loc[rating].sort_values(ascending=False)[0:20]/T.loc[rating]).plot(kind='barh',
                                                                          ax=ax,
                                                                          title=f'Most frequent stems {int(rating)} stars \n average frequencies',
                                                                         )
plt.tight_layout()
```

```python
A = df_stems.groupby('stars_rat').sum()
```

```python
B = A.sum(axis=1) / df_text['stars_rating'].value_counts().sort_index()
```

```python
sns.barplot(x = B.index ,y = B.values)
plt.title('Average number of stems per review')
```

## Term frequency matrix

### per review

Divide number of occurence in a review by number of words in a review

```python
df_stems_freq = df_stems.iloc[:,2:].divide(df_stems.iloc[:,2:].sum(axis=1)[:],axis=0)
```

```python
df_stems_freq
```

### per star rating

```python
df_stem_rat = df_stems.drop(columns='text').groupby('stars_rat').sum()
```

```python
df_stem_rat
```

```python
df_stem_rat.sum(axis=1)[:].sum()
```

```python
df_stem_rat.sum(axis=1)[:]
```

```python
df_stem_rat_freq = df_stem_rat.divide(df_stem_rat.sum(axis=1)[:],axis=0)
```

```python
df_stem_rat_freq
```

## IDF matrix

### per review

number of reviews with stem in it

```python
reviews_per_stem = df_stems.iloc[:,2:].replace(to_replace=0,value=np.nan).count()
```

```python
reviews_per_stem
```

```python
idf_matrix =np.log10(df_stems.shape[0] / reviews_per_stem)
```

```python
idf_matrix
```

### per star rating

```python
df_stem_rat
```

```python
ratings_per_stem =  df_stem_rat.replace(to_replace=0,value=np.nan).count()
```

```python
ratings_per_stem
```

```python
df_stem_rat.shape[0]
```

```python
idf_matrix_rat = np.log10(df_stem_rat.shape[0] / ratings_per_stem)
```
